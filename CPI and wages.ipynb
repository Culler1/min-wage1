```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
series_id=pd.read_table("https://download.bls.gov/pub/time.series/ap/ap.series")
```

```python
food1=pd.read_table("https://download.bls.gov/pub/time.series/ap/ap.data.3.Food")
food1.columns=food1.columns.str.strip()
food1['series_id']=food1.series_id.str.strip()
food1.rename(columns={'period':'month','value':'price'},inplace=True)
food1['price']=food1.price.str.strip().str.replace("-","0").astype(float).round(2)
food1['month']=food1.month.str.replace('M','').astype(int)
food1.sort_values(by=['year','month'],inplace=True)
```

```python
#milk1=food1[food1.series_id=="APU0000709112"]
#Average price of milk per gallon in northeast urban areas.
milk1=food1[food1.series_id=="APU0100709112"]
```

```python
milk1.groupby(['year','month']).price.sum().plot(kind='line')
```
